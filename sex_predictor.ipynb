Vamos constuir nosso modelo
NÃ³s podemos utilizar todos dados do nosso dataset e criar um modelo de machine learning pronto para receber novos dados e realizar suas previsÃµes, sÃ³ que com isso nunca irÃ­amos saber o real desempenho do nosso modelo. O algoritmo poderia aprender perfeitamente a relaÃ§Ã£o existentes nos dados apresentados e com isso criar um modelo que sofre de Overfitting e sÃ³ descobrirÃ­amos esse problema apÃ³s as previsÃµes desastrosas geradas por este modelo.

Para medir o real desempenho de um modelo, Ã© fundamental que realizemos testes com ele, utilizando dados diferentes dos que foram apresentados em sua criaÃ§Ã£o. Assim, apÃ³s a realizaÃ§Ã£o do prÃ©-processamento, iremos separar a totalidade dos dados em dois grupos, sendo o primeiro responsÃ¡vel pelo aprendizado do modelo, e o segundo por realizar os testes.



```python
#aqui importamos as funÃ§Ãµes do scikit para proceder com nossa #anÃ¡lise
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#importamos aqui o seaborn para analisar os dados
import seaborn as sns
sns.set()
```

```python
#aqui iremos selecionar 4 colunas para nossa anÃ¡lise
X = dff[['hc','restecg','thalach']]
```

Esses  dados sÃ£o significativos para determinarmos o sexo

```python
y=dff[['sex']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
#Por padrÃ£o, train_test_split reserva 25% das amostras no conjunto de dados original para teste.
```

Isso nos dÃ¡ dois conjuntos de dados â€” um para treinamento e outro para teste. Vamos treinar o modelo.

Para fazer uso do algoritmo KNN, Ã© suficiente criar uma instÃ¢ncia de KNeighborsClassifier. Por padrÃ£o, o KNeighborsClassifier procura os 5 vizinhos mais prÃ³ximos, por isso temos n_neighbors=5, a mÃ©trica utilizada aqui Ã© a distÃ¢ncia euclidiana, metric=â€™euclideanâ€™.

```python
knn = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

Uma forma de avaliar nosso modelo Ã© calcular a matriz de confusÃ£o. Os nÃºmeros na diagonal da matriz de confusÃ£o correspondem Ã s previsÃµes corretas, enquanto os outros implicam em falsos positivos e falsos negativos.



```python
from sklearn.metrics import classification_report, confusion_matrix
cm=np.array(confusion_matrix(y_test,y_pred, labels=[1,0]))
confusion=pd.DataFrame(cm, index=['masculino', 'feminino'], columns=['previu que era masculino', 'previu que era feminino'])
confusion
```

```python
 accuracy = knn.score(X_test,y_test)
```

```python
accuracy
```

Obtivemos aqui uma acurÃ¡cia de 62,5% para nosso modelo.
Vamos procurar ver uma forma de melhorar esse valor

O cÃ³digo a seguir repete o experimento 2000 vezes e faz um histograma das acurÃ¡cias obtidas durante essas repetiÃ§Ãµes.
Faremos um experimento para escolher o melhor nÃºmero de vizinhos, por meio do cÃ³digo abaixo:




```python
mean_scores = []
for k in range(1,10):
    scores = []
    for i in range(60):
        X_train, X_test, y_train, y_test = train_test_split(X,y)
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_train,y_train)
        accuracy = model.score(X_test,y_test)
        scores.append(accuracy)
    mean_scores.append(np.mean(scores))

import matplotlib.pyplot as plt
import seaborn as sns
plt.plot(np.arange(1,10),mean_scores)
plt.yticks([])
plt.title("AcurÃ¡cias do k-NN por nÃºmero de vizinhos")
plt.xlabel("NÃºmero de vizinhos")
plt.ylabel("AcurÃ¡cia")
plt.show()
```

No cÃ³digo acima, variamos o nÃºmero de vizinhos k de 1 a 10 ( linha 19). Repetimos 60 vezes (linha 21) para cada k, o treinamento e a validaÃ§Ã£o. Na lista mean_scores salvamos a acurÃ¡cia mÃ©dia obtida para cada k e por fim, mostramos um grÃ¡fico de linha de k pela acurÃ¡cia mÃ©dia do modelo k-NN com o nÃºmero de vizinhos igual a k.

para altos valores de k nosso modelo perde acurÃ¡cia. O nÃºmero de vizinho padrÃ£o, igual a 2, obteve melhor desempenho nesse experimento!

Normalizando os dados
Em alguns problemas de classificaÃ§Ã£o e, dependendo do algoritmo usado, Ã© importante normalizar os dados.

Vamos a mais um experimento:

```python
from sklearn.preprocessing import Normalizer
# normalizando
scaler = Normalizer()
scaler.fit(X)
X = scaler.transform(X)

scores = []
for i in range(2000):
    X_train, X_test, y_train, y_test = train_test_split(X,y)
    model = KNeighborsClassifier()
    model.fit(X_train,y_train)
    accuracy = model.score(X_test,y_test)
    scores.append(accuracy)
    
print("MÃ©dia: {:.2f}%".format(np.mean(scores)*100))
print("Desvio padrÃ£o: {:.2f}%".format(np.std(scores)*100))

import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(scores)
plt.yticks([])
plt.title("AcurÃ¡cias do k-NN")
plt.show()
```

Note que nÃ£o foi muito diferente dos resultados obtidos sem normalizar.
