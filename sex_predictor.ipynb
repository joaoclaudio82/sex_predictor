Vamos constuir nosso modelo
Nós podemos utilizar todos dados do nosso dataset e criar um modelo de machine learning pronto para receber novos dados e realizar suas previsões, só que com isso nunca iríamos saber o real desempenho do nosso modelo. O algoritmo poderia aprender perfeitamente a relação existentes nos dados apresentados e com isso criar um modelo que sofre de Overfitting e só descobriríamos esse problema após as previsões desastrosas geradas por este modelo.

Para medir o real desempenho de um modelo, é fundamental que realizemos testes com ele, utilizando dados diferentes dos que foram apresentados em sua criação. Assim, após a realização do pré-processamento, iremos separar a totalidade dos dados em dois grupos, sendo o primeiro responsável pelo aprendizado do modelo, e o segundo por realizar os testes.



```python
#aqui importamos as funções do scikit para proceder com nossa #análise
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#importamos aqui o seaborn para analisar os dados
import seaborn as sns
sns.set()
```

```python
#aqui iremos selecionar 4 colunas para nossa análise
X = dff[['hc','restecg','thalach']]
```

Esses  dados são significativos para determinarmos o sexo

```python
y=dff[['sex']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
#Por padrão, train_test_split reserva 25% das amostras no conjunto de dados original para teste.
```

Isso nos dá dois conjuntos de dados — um para treinamento e outro para teste. Vamos treinar o modelo.

Para fazer uso do algoritmo KNN, é suficiente criar uma instância de KNeighborsClassifier. Por padrão, o KNeighborsClassifier procura os 5 vizinhos mais próximos, por isso temos n_neighbors=5, a métrica utilizada aqui é a distância euclidiana, metric=’euclidean’.

```python
knn = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

Uma forma de avaliar nosso modelo é calcular a matriz de confusão. Os números na diagonal da matriz de confusão correspondem às previsões corretas, enquanto os outros implicam em falsos positivos e falsos negativos.



```python
from sklearn.metrics import classification_report, confusion_matrix
cm=np.array(confusion_matrix(y_test,y_pred, labels=[1,0]))
confusion=pd.DataFrame(cm, index=['masculino', 'feminino'], columns=['previu que era masculino', 'previu que era feminino'])
confusion
```

```python
 accuracy = knn.score(X_test,y_test)
```

```python
accuracy
```

Obtivemos aqui uma acurácia de 62,5% para nosso modelo.
Vamos procurar ver uma forma de melhorar esse valor

O código a seguir repete o experimento 2000 vezes e faz um histograma das acurácias obtidas durante essas repetições.
Faremos um experimento para escolher o melhor número de vizinhos, por meio do código abaixo:




```python
mean_scores = []
for k in range(1,10):
    scores = []
    for i in range(60):
        X_train, X_test, y_train, y_test = train_test_split(X,y)
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_train,y_train)
        accuracy = model.score(X_test,y_test)
        scores.append(accuracy)
    mean_scores.append(np.mean(scores))

import matplotlib.pyplot as plt
import seaborn as sns
plt.plot(np.arange(1,10),mean_scores)
plt.yticks([])
plt.title("Acurácias do k-NN por número de vizinhos")
plt.xlabel("Número de vizinhos")
plt.ylabel("Acurácia")
plt.show()
```

No código acima, variamos o número de vizinhos k de 1 a 10 ( linha 19). Repetimos 60 vezes (linha 21) para cada k, o treinamento e a validação. Na lista mean_scores salvamos a acurácia média obtida para cada k e por fim, mostramos um gráfico de linha de k pela acurácia média do modelo k-NN com o número de vizinhos igual a k.

para altos valores de k nosso modelo perde acurácia. O número de vizinho padrão, igual a 2, obteve melhor desempenho nesse experimento!

Normalizando os dados
Em alguns problemas de classificação e, dependendo do algoritmo usado, é importante normalizar os dados.

Vamos a mais um experimento:

```python
from sklearn.preprocessing import Normalizer
# normalizando
scaler = Normalizer()
scaler.fit(X)
X = scaler.transform(X)

scores = []
for i in range(2000):
    X_train, X_test, y_train, y_test = train_test_split(X,y)
    model = KNeighborsClassifier()
    model.fit(X_train,y_train)
    accuracy = model.score(X_test,y_test)
    scores.append(accuracy)
    
print("Média: {:.2f}%".format(np.mean(scores)*100))
print("Desvio padrão: {:.2f}%".format(np.std(scores)*100))

import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(scores)
plt.yticks([])
plt.title("Acurácias do k-NN")
plt.show()
```

Note que não foi muito diferente dos resultados obtidos sem normalizar.
